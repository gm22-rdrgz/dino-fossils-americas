## Confirm changes

```python
df = df_drop
```

Rename columns into readable names

```python
df_renamed = df.rename(columns={
    'accepted_name': 'Accepted Taxon Name', 'accepted_no': 'Accepted Taxon ID', 
    'early_interval': 'Early Interval', 'late_interval': 'Late Interva', 
    'max_ma': 'Maximum Age (Ma)', 'min_ma': 'Minimum Age (Ma)', 
    'phylum': 'Phylum', 'class': 'Class', 'family': 'Family', 
    'genus': 'Genus', 'abund_value': 'Abundance Value', 
    'abund_unit': 'Abundance Unit', 'lng': 'Longitude', 
    'lat': 'Latitude', 'cc': 'Country Code', 
    'state': 'State', 'county': 'County',
    'latlng_precision': 'Coordinate Precision',
    'altitude_value': 'Altitude',
    'altitude_unit': 'Altitude Unit',
    'geogscale': 'Geographic Scale',
    'geogcomments': 'Location Comments', 
    'paleoage': 'Geologic Interval',
    'collection_dates': 'Collection Year',
    'diet': 'Diet'})
df_renamed
```

## Confirm Changes

```python
df = df_renamed
```

## Transform taxon name and diet into title case

```python
df["Accepted Taxon Name"] = df["Accepted Taxon Name"].str.strip().str.title()
df["Diet"] = df["Diet"].str.strip().str.title()
```

## Change "Nan" values to a more context-appropriate value

```python
df['Genus'] = df['Genus'].fillna("Unknown")
df['Diet'] = df['Diet'].fillna("Unknown")
df['Collection Year'] = df['Collection Year'].fillna("No data")
```

## Write the transformed dinosaur data to CSV
